```python
import numpy as np
```

```python
arr = np.array([1,2,3])
arr = arr.astype(np.int8)
print(arr.shape)
print(arr.size)
print(arr.ndim)
print(arr.dtype)
print(type(arr))
```

```python
#zero
arr1 = np.zeros((2,3), dtype=np.int64)
#print(arr1)

arr2 = np.ones((3,2))
# print(arr2)

arr3 = np.empty((2,2),dtype=np.int64)
# print(arr3)

arr4 = np.full((2,3), np.inf)
print(arr4)
```

```python
arr5 = np.diag((3,4,5))
# print(arr5)

arr6 = np.eye((3), dtype= np.int64)
print(arr6)
```

### dictionary

```python
dit = { 
    'a' : 4,
    'b' : 5,
    'c' : 8,
    'e' : 6
}

arr_val = np.array(list(dit.values()))
print(arr_val)

arr_key = np.array(list(dit.keys()))
print(arr_key)
```

#### array with random value

```python
arr = np.random.rand(2,3)
print(arr)
print(arr.shape)
print(arr.size)
print(arr.dtype)
print(arr.ndim)
```

```python
arr = np.random.randint(1,10,(3,3,2))
print(arr)
```

```python
                               #shape, row, column
arr = np.random.uniform(40,80, (3,2,3))
print(arr)
```

#### array creation with range functions

```python
# np.random.uniform(start,end, shape) 
arr = np.arange(1,10, 1)
mat = arr.reshape(3,3)
print(mat)
```

```python
arr = np.linspace(1,6 , 10)
print(arr)

arr1 = np.logspace(1,6 , 4)
print(arr1)
```

### indexing and slicing

```python
arr = np.array([1,2,3,4])

print(arr[2])

arr1 = np.array([[1,2,3],[2,3,5]])
print(arr1[1][2])
```

```python
# slicing
arr_mod = arr[2:]
print(arr_mod)
```

```python
arr1_mod = arr1[0:2,0:2]
print(arr1_mod)
```
