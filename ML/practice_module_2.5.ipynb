2. Temperatures (°C): 29, 31, 33, 33, 32, 31, 30

Calculate variance and standard deviation.

Which measure is easier to interpret and why?


```python
Marks = pd.DataFrame({'mark':[50, 60, 65, 70, 75, 80, 85, 90]})

Temperatures = pd.DataFrame({'temp':[29, 31, 33, 33, 32, 31, 30]})
```

```python
print('mean: ',np.mean(Marks))
print('median: ',np.median(Marks))
Marks['mark'].mode()
```

```python
print('varriance: ',Temperatures['temp'].var())
print('std: ',Temperatures['temp'].std())
```

3. Cat weights (kg): 2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5

Find P25, P50, P75 and IQR. Mark possible outliers.

4. Given μ = 50, σ = 10

Find Z for x = 65 and interpret.

For Z = −2.5, find x.


```python
df = pd.DataFrame({'weights' :[2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5]})

p25 = df['weights'].quantile(.25)
p50 = df['weights'].quantile(.50)
p75 = df['weights'].quantile(.75)

print(p25, p50, p75)

IQR = p75 - p25
print(IQR)
```
