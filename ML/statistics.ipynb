```python
data = [120, 130, 125, 140, 135, 200, 220]
df = pd.DataFrame({'Purchase': data})

print('mean',df['Purchase'].mean())

pop_var = np.var(df['Purchase'],ddof=0)
samp_var = np.var(df['Purchase'],ddof=1)
pop_std = np.std(df['Purchase'],ddof=0)
samp_std = np.std(df['Purchase'],ddof=1)


print('pop_var:',pop_var)
print('samp_var:',samp_var)
print('pop_std:',samp_var)
print('samp_std:',samp_std)
```

```python
data = [10, 12, 13, 15, 18, 21, 23, 25, 28, 30, 34, 35, 37, 40]

print('Q1: ',np.percentile(data, 25))
print('Q2: ',np.percentile(data, 50))
print('Q3: ',np.percentile(data, 75))
```

```python
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)

IQR = Q3 - Q1

lower_bound = Q1-(1.5*IQR)
uper_bound = Q3+(1.5*IQR)

print('IQR: ',IQR)
print('Quantile range: ',lower_bound ,'to', uper_bound)
```

```python
men = np.mean(data)
std = np.std(data)

z_score = [(x-men)/ std for x in data]
print('Z_score: ',np.round(z_score,2))
```

```python
df = pd.DataFrame({'Income': [22, 25, 27, 29, 35, 40, 42, 100, 110, 115]})
std = np.std(df['Income'])
var = np.var(df['Income'])
mean = df['Income'].mean()

Q1 = df['Income'].quantile(.25)
Q3 = df['Income'].quantile(.75)

IQR = Q3-Q1

lower = Q1 - (1.5*IQR)
uper = Q3 + (1.5*IQR)

outliers_iqr = df[(df['Income']<lower) | (df['Income']>uper)]
print("Outliers iqr: ",outliers_iqr)

df['z_score'] = (df['Income'] - mean)/std
outliers_z = df[np.abs(df['z_score']>2.5)]
print('outliers z',outliers_z)


median = df['Income'].median()
mad = np.median(np.abs(df['Income']-median))
df['mod_z'] = (df['Income']-median)/mad
outliers_mz = df[np.abs(df['mod_z']>3.5)]
outliers_mz
```
