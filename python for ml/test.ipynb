```python
n = int(input())
target = float(input())

sums = 0.0

for i in range(n):
    loss = float(input())
    sums += loss

average_loss = sums/n 

if average_loss <= target:
    print('PASS')
else:
    print('RETRY')
  
    
```

```python
n = int(input())

yes_count = 0
no_count = 0
for i in range(n):
    vote = input().upper()

    if vote == 'YES':
        yes_count +=1
    elif vote == 'NO':
        no_count +=1 

if yes_count>= no_count:
    print('ACCEPT')
else:
    print('REJECT')
```

```python
pixel = input().split()
brightness  = []
for i in pixel:
    brightness.append(int(i))
s = 0
for i in brightness:
    s += i
average_brightness = s/len(brightness)

if average_brightness < 85:
    print('Dark Image')
elif 85 <= average_brightness <= 170:
    print('Normal Image')
elif average_brightness > 170:
    print('Bright Image')
    
```

```python
x = input().split()

a = x.count('A')
b = x.count('B')


A = (a/len(x))*100
B = (b/len(x))*100

if A > 70 or B >70:
    print('Biased Model')
else:
    print('Fair Model')
```
