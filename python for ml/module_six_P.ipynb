##### Lambda

### square

```python
lst = [1,2,3,4]

squre = list(map(lambda x:x**2, lst))
print(squre)
```

### Even


```python
lst = [x for x in range(100)]

even = list(filter(lambda x:x%2==0 , lst))
print(even)
```

```python
lst = [x for x in range(100)]

even = lambda x,y: x*y

print(even(2,3))
```

```python
words = ["apple", "banana", "avocado", "mango"]

# Filter words starting with 'a'
result = tuple(filter(lambda word: word.startswith('a'), words))

print(result)
```
