```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('practice.csv')
```

1. Add New Column and Save File
You have a DataFrame with column Price.
 Create a new column Discounted_Price with (10% discounted price).
 Save the modified DataFrame as discounted_products.csv.

```python
new_df = df.copy()
new_df['Discounted_Price'] = (new_df['Price'] - (new_df['Price']*.1))
new_df.to_csv('discounted_prodicts.csv')
new_df
```

2. Check Duplicates and Remove Them
Column: EmployeeID
 Tasks:
Count duplicate entries


Remove duplicates


Display cleaned DataFrame


```python
print(df['EmployeeID'].duplicated().sum())
df.drop_duplicates(subset='EmployeeID',inplace=True)
df
```

3. Handle Missing Values
Columns: Age, Salary
 Tasks:
Count missing values


Fill missing Age with mean age


Drop rows where Salary is missing

```python
print(df.isnull().sum())
df['Age'].fillna(df['Age'].mean(),inplace=True)
df.dropna(subset='Salary',inplace=True)
df
```

4. Apply Statistical Functions
Columns: Marks1, Marks2, Marks3
 Tasks:
Calculate mean, max for each subject


Compute total marks per student


Find standard deviation of total marks


```python
print('Mean:',df['Marks1'].mean())
print('Max:',df['Marks1'].max())

print('Mean:',df['Marks2'].mean())
print('Max:',df['Marks2'].max())

print('Mean:',df['Marks3'].mean())
print('Max:',df['Marks3'].max())
```

```python
df['Total marks'] = df['Marks1'] + df['Marks2'] + df['Marks3']
print(df)
df['Total marks'].std()
```

5. Histogram, Bar, Scatter, Pie
Create:
Histogram of Age
Scatter plot (Age vs Salary)
Scatter plot ( Price vs Discounted Price) 


```python
plt.title('Histogram of age column')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.hist(df['Age'])
plt.show()
```

```python
plt.title('scatter plot')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(df['Age'],df['Salary'])
plt.show()
```

```python
plt.title('Scatter plot price and discount')
plt.xlabel('price')
plt.ylabel('discount')
plt.scatter(new_df['Price'],new_df['Discounted_Price'])
plt.show()
```
