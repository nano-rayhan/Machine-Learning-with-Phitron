```python
# You must run this code and use that data set for all the questions
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

### Q1. Create NumPy arrays from the ages and salaries data generated above. Print both arrays and display their dtype, ndim, shape, and size.

```python
ages_array = np.array(ages)
salaries_array = np.array(salaries)

print('ages_array: ',ages_array)
print('dtype: ',ages_array.dtype)
print('ndim: ',ages_array.ndim)
print('shape: ',ages_array.shape)
print('size: ',ages_array.size)

print('salaries_array: ',salaries_array)
print('dtype: ',salaries_array.dtype)
print('ndim: ',salaries_array.ndim)
print('shape: ',salaries_array.shape)
print('size: ',salaries_array.size)
```

### Q2. Using the salaries array, find and print the highest and lowest salary values. Also, calculate and print the average salary and age using NumPy functions.

```python
print(salaries_array)
highest_salary = salaries_array.max()
print('Highest Salaries: ',highest_salary)
lowest_slary = salaries_array.min()
print('Lowest Salary: ',lowest_slary)
print('Average salary: ',salaries_array.mean())
print('Average age: ',ages.mean())
```

### Q3. From the ages array, filter and print all ages greater than 30 using a boolean condition. Then count and print how many employees are older than 30.

```python
bool_age = ages_array>30
greater_than_30 = ages_array[bool_age]
print('greater_than_30: ',greater_than_30)
print('Employees are older than 30: ',len(greater_than_30))
```

### Q4. Create a new NumPy array that increases every employeeâ€™s age by 5 years (without modifying the original array). Print the new updated ages array.

```python
print(ages_array)
updated_age_array = ages_array + 5
print('After increase 5 years: ',updated_age_array)
```

### Q5. Using NumPy, calculate the total salary expense (sum of all salaries) and the difference between the maximum and minimum salary. Print both results.

```python
print('Salary expense: ',salaries_array.sum())
max_salary = salaries_array.max()
min_salary = salaries_array.min()
diference = max_salary - min_salary
print('max_salary - min_salary: ',diference)
```
