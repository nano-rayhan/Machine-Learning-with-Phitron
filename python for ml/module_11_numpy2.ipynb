```python
import numpy as np
```

### Flatten colum or row wise

```python
arr = np.random.randint(1,100, (4,5))
print(arr)
falten_arr = np.ravel(arr, order='A')
print(falten_arr)
```

### concanate

```python
a = np.random.randint(1,11,(2,3))
b = np.random.randint(1,11, (2,3))
print(a)
print(b)

concatenate = np.concat((a,b), axis=1)

print(concatenate)
```

```python
# transpose matrix

mat = np.random.randint(1,11, (2,3))

print(mat.T.shape)
mat.shape
```

### search array

```python
arr = np.array([[1,2,3,4],
                [3,5,6,7]])

index = np.where(arr==10)
print(index)
```

### count array

```python
arr = np.random.randint(1,100,(100,))
print(arr)

count_number = np.count_nonzero(arr<40)
# print(count_number)

unique_value = np.unique(arr, return_counts=True)
print(unique_value)
```
