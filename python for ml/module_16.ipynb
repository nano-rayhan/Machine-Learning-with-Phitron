```python
students = pd.read_csv('student_dataset_complete.csv')
```

```python
sns.scatterplot(data=students, x='study_hours',y='test_score',hue='gender')
```

```python
sns.relplot(kind='scatter',data=students, x='study_hours',y='test_score',hue='gender',col='week',col_wrap=2)
```

## Scatter plot with plotly

```python
fig = px.scatter(students, x='time_study',y='Marks',
                 color='gender',
                 size='Tshirt_size',
                 hover_data=['hostel'])
fig.show()
```
