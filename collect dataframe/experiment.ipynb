## Web Scraping demo

```python
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36"
}

res =  requests.get('https://en.wikipedia.org/wiki/List_of_companies_of_Bangladesh',headers=headers).text
```

```python
soup = BeautifulSoup(res, 'lxml')
```

```python
print(soup.prettify())
```

```python
table = soup.find('table')
```

```python
table
```

```python
tr = table.find_all('tr')
tr
```

```python
tr[4]
```

```python
company = []
industry = []
headquaters = []
founded = []
for t in tr[2:]:
    td = t.find_all('td')
    company.append(td[0].text.strip())
    industry.append(td[1].text.strip())
    headquaters.append(td[3].text.strip())
    founded.append(td[4].text.strip())
```

```python
df = pd.DataFrame(
    {
        'company' : company,
        'industry' : industry,
        'headquaters' : headquaters,
        'founded' : founded,
    }
)
```
