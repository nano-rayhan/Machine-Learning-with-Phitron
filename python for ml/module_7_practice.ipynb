### Problem 1: Common Friends (Set Operations)


```python
a_friends = {"Rahim", "Karim", "Sakib", "Jamal"}
b_friends = {"Sakib", "Jamal", "Rafiq", "Nadim"}

all_friends = list(a_friends | b_friends)

mutual_friends = set(filter(lambda x: x in a_friends and x in b_friends, all_friends))

unique_a = set(filter(lambda x: x in a_friends and x not in b_friends, all_friends))
unique_b = set(filter(lambda x: x in b_friends and x not in a_friends, all_friends))

number_of_unique = set(filter(lambda x: x in a_friends or x  in b_friends, all_friends))

print(f'Mutual friends: {mutual_friends}')
print(f'Unique to A: {unique_a}')
print(f'Unique to B: {unique_b}')
print(f'Total unique friends: {len(number_of_unique)}')
```

### Problem 2: Sort Students by Marks (Tuple + Dictionary)

```python
students = [("Rafi", 89), ("Sumi", 95), ("Hasan", 90), ("Nila", 75), ("Anik", 98)]

sorted_list = sorted(students, key= lambda x: x[1], reverse=True)

top_3 = [x for x in sorted_list[:3]  ]
print("Top 3 students:")
for name, mark in top_3:
    print(f'{name} - {mark}')
```

### Problem 3: Word Length Filter (Lambda + Filter)

```python
words = ["sun", "planet", "moon", "star", "universe"]

greter_than_4 = list(filter(lambda x: len(x)>4 , words))

print(greter_than_4)
```

### Problem 4: Square of Even Numbers (Map + Filter + Lambda)


```python
nums = [1, 2, 3, 4, 5, 6]

even_square = list(map(lambda x :x**2 ,filter(lambda x : x%2==0, nums)))
print(even_square)
```
