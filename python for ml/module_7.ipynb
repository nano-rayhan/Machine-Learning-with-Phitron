#### Read File

```python
file = open('sample.txt','r')

content = file.read()

print(content)
file.close()
```

#### Read Lines

```python
file = open('sample.txt','r')

content = file.readlines()

print(content)
file.close()
```

```python
file = open('sample.txt','r')

content = file.read()
content.strip()

print(content)
file.close()
```
