### creating class and object

```python
class Phone:
    category = 'Electronics'

    def __init__(self, model, camera, battary, battary_percentage=100):
        self.model = model
        self.camera = camera
        self.battary = battary
        self.battary_percentage = battary_percentage

    def charge(self,hour):
        print(f'Charge completed by {hour}')
        return 

class Cooling_mechanism:
    def __init__(self, cooling_method):
        self.cooling_method = cooling_method

    def cooling_on(self):
        print(f"the system is being cool by {self.cooling_method}")
         

iPhone = Phone('iPhone 17', 108, 5000)
print(iPhone.charge(2))

coll = Cooling_mechanism('heat')
print(coll.cooling_on())
```

### Single inheritnace

```python
class SmartPhone(Phone):
    def __init__(self, model, camera, battary, battary_percentage=100):
        super().__init__(model, camera, battary, battary_percentage)
    
    def charge(self, hour):
        print('Fast charges process ')
        super().charge(hour)

Redmi = SmartPhone('Redmi 13', '108', 5500)
print(Redmi.charge(1))
```

### Multiple inheritence

```python
class SmartPhone_Cooling(SmartPhone, Cooling_mechanism):
    def __init__(self, model, camera, battary, battary_percentage=100):
        SmartPhone.__init__(self, model, camera, battary, battary_percentage)
        Cooling_mechanism.__init__(self, cooling_method=self)

pro_cooling = SmartPhone_Cooling('iPhone', 100, 5500)
 # SmartPhone class theke
print(pro_cooling.battary) # Phone class theke
print(pro_cooling.cooling_method) # Cooling_Mechanism class theke
print(pro_cooling.cooling_on()) # Cooling_Mechanism class er method
print(pro_cooling.charge(1))

```
