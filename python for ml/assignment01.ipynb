### Using filter() and lambda, extract even numbers from a given list.

```python
numbers = input().split()
num = list(map(int, numbers))
evne_num = list(filter(lambda x: x%2==0, num))
print(evne_num)
```

### Write a function safe_divide(a,b) that performs a/b and handles division by zero using tryâ€“except.
 Print "Division by zero" if b = 0.

```python
a = input().split()

def safe_devide(a, b):
    return a/b

try:
    print(safe_devide(int(a[0]), int(a[1])))

except ZeroDivisionError:
    print('Division by zero')
```

### Find and print the longest word in a given sentence (if tie, print first one).

```python

from functools import reduce


text = input()

word = text.split()

longest = reduce(lambda x, y: x if len(x)>= len(y) else y, word)
print(longest)
```

### A number is called strong if the sum of the factorial of its digits equals the number itself.
 Example: 145 = 1! + 4! + 5!

```python
n = int(input())

real_num = n

digits = []

while n > 0:
    val = n%10
    digits.insert(0 , val)
    n //= 10
    
def fact(x):
    if x == 0 or x == 1:
        return 1
    else:
        return x*fact(x-1)
    
num = list(map(int, digits))
sums = 0
for val in num:
    f = fact(val)
    sums +=  f

if real_num == sums:
    print('Strong Number')
else:
    print('Not Strong Number')
```
